# Merge and Join in Pandas 🔗

Real-world data is often split across multiple tables.
Merge and join help combine datasets into one.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
```

## Sample Student Data


```python
students = pd.DataFrame({
    "StudentID": [1, 2, 3, 4],
    "Name": ["Abhay", "Rahul", "Aman", "Neha"]
})

marks = pd.DataFrame({
    "StudentID": [1, 2, 4],
    "Marks": [85, 90, 95]
})

print(students)
print(marks)
```

## Merge (Inner Join)


```python
inner = pd.merge(students, marks, on="StudentID", how="inner")
print(inner)
```

## Left Join


```python
left = pd.merge(students, marks, on="StudentID", how="left")
print(left)
```

## Right Join


```python
right = pd.merge(students, marks, on="StudentID", how="right")
print(right)
```

## Outer Join


```python
outer = pd.merge(students, marks, on="StudentID", how="outer")
print(outer)
```

## Joining Using Index


```python
students_indexed = students.set_index("StudentID")
marks_indexed = marks.set_index("StudentID")

joined = students_indexed.join(marks_indexed)
print(joined)
```

## Why Merge & Join are Important?


- Combine datasets
- Create training tables
- Join features from multiple sources
