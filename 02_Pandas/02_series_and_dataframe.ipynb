# Pandas Series and DataFrame 📊

Series and DataFrame are the two main data structures in Pandas.
Understanding them is critical for data analysis and machine learning.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
```

## Pandas Series

A Series is a one-dimensional labeled array.
It is similar to a column in Excel.


```python
s = pd.Series([10, 20, 30, 40])
print(s)
```

## Series with Custom Index


```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
```

## Accessing Series Elements


```python
print(s["a"])
print(s[1])
```

## Pandas DataFrame

A DataFrame is a 2-dimensional table with rows and columns.


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman"],
    "Age": [20, 21, 22],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

## Accessing Columns


```python
print(df["Name"])
print(df["Marks"])
```

## Accessing Rows


```python
print(df.loc[0])
print(df.iloc[1])
```

## Adding a New Column


```python
df["Passed"] = df["Marks"] >= 40
print(df)
```

## Modifying Data


```python
df.loc[0, "Marks"] = 95
print(df)
```

## Deleting Columns


```python
df.drop("Passed", axis=1, inplace=True)
print(df)
```

## Why Series & DataFrame Matter in ML?


- ML models accept tabular data
- Features are columns
- Records are rows
- Pandas prepares data for training
