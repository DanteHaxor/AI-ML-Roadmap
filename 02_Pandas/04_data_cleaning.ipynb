# Data Cleaning in Pandas 🧹

Real-world data is messy.
Before training any Machine Learning model,
data must be cleaned and prepared.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
import numpy as np
```

## Sample Messy Dataset


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman", "Neha", "Rahul"],
    "Age": [20, np.nan, 22, 19, np.nan],
    "Marks": [85, 90, np.nan, 92, 90]
}

df = pd.DataFrame(data)
print(df)
```

## Detecting Missing Values


```python
print(df.isnull())
print(df.isnull().sum())
```

## Filling Missing Values


```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
print(df)
```

## Dropping Missing Values


```python
df_dropped = df.dropna()
print(df_dropped)
```

## Removing Duplicates


```python
print(df.duplicated())
df = df.drop_duplicates()
print(df)
```

## Renaming Columns


```python
df.rename(columns={"Marks": "Score"}, inplace=True)
print(df)
```

## Changing Data Types


```python
df["Age"] = df["Age"].astype(int)
print(df.dtypes)
```

## Why Data Cleaning is Critical?


- ML models cannot handle missing values
- Duplicate data causes bias
- Clean data = better accuracy
