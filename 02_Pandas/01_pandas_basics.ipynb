# Pandas Basics 🐼

Pandas is a powerful Python library for data manipulation and analysis.
It is used to handle structured data like Excel sheets, CSV files, and databases.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
```

## Why Pandas?

Pandas helps to:
- Load large datasets
- Clean messy data
- Analyze and filter data
- Prepare data for Machine Learning


## Creating a Series


```python
data = [10, 20, 30, 40]
s = pd.Series(data)

print(s)
```

## Creating a DataFrame


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman"],
    "Age": [20, 21, 22],
    "Course": ["AI/ML", "DS", "AI"]
}

df = pd.DataFrame(data)
print(df)
```

## DataFrame Information


```python
print(df.shape)
print(df.columns)
print(df.dtypes)
```

## Head and Tail


```python
print(df.head())
print(df.tail())
```

## Basic Statistics


```python
print(df.describe())
```

## Why Pandas is Important for ML?


Pandas helps in:
- Cleaning missing values
- Feature engineering
- Filtering rows
- Selecting columns
- Preparing training data
