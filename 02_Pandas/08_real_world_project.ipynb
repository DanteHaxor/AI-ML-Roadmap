```python
import pandas as pd
import numpy as np
```

## Step 1 – Create a Realistic Dataset


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman", "Neha", "Riya", "Aman"],
    "Math": [85, 90, 78, 92, np.nan, 78],
    "Science": [88, 85, 80, 95, 89, 80],
    "English": [90, 87, 75, 88, 92, 75]
}

df = pd.DataFrame(data)
print(df)
```

## Step 2 – Detect Missing Values


```python
print(df.isnull().sum())
```

## Step 3 – Fill Missing Values


```python
df["Math"].fillna(df["Math"].mean(), inplace=True)
print(df)
```

## Step 4 – Remove Duplicate Records


```python
df = df.drop_duplicates()
print(df)
```

## Step 5 – Calculate Average Marks


```python
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)
print(df)
```

## Step 6 – Assign Grades


```python
def grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    else:
        return "D"

df["Grade"] = df["Average"].apply(grade)
print(df)
```

## Step 7 – Find Top Performers


```python
top_students = df[df["Average"] >= 85]
print(top_students)
```

## Step 8 – Prepare ML Ready Data


```python
features = df[["Math", "Science", "English"]]
labels = df["Grade"]

print("Features:\n", features)
print("Labels:\n", labels)
```

## Why This is Important for ML?


This pipeline:
- Cleans data
- Removes noise
- Creates features
- Prepares labels

This is exactly what happens before training any ML model.
