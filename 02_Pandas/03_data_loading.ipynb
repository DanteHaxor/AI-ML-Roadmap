# Loading Data in Pandas 📂

In real life, data comes from files like CSV, Excel, and databases.
Pandas allows us to easily load and explore such data.

```python
import pandas as pd
```

## Loading CSV Files


```python
# Example: df = pd.read_csv("data.csv")
```

## Creating a Sample CSV Using Pandas


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman", "Neha"],
    "Age": [20, 21, 22, 19],
    "Marks": [85, 90, 88, 92]
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)
print("students.csv created")
```

## Reading CSV File


```python
df = pd.read_csv("students.csv")
print(df)
```

## Viewing Dataset


```python
print(df.head())
print(df.tail())
```

## Dataset Information


```python
print(df.info())
print(df.describe())
```

## Column Selection


```python
print(df["Name"])
print(df[["Name", "Marks"]])
```
