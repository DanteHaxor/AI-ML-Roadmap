# GroupBy and Aggregation in Pandas 📊

GroupBy is used to split data into groups
and apply operations like sum, mean, count, etc.

This is one of the most important concepts in data analysis.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
```

## Sample Sales Dataset


```python
data = {
    "City": ["Delhi", "Delhi", "Mumbai", "Mumbai", "Pune", "Pune"],
    "Product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "Sales": [50000, 30000, 55000, 28000, 48000, 25000]
}

df = pd.DataFrame(data)
print(df)
```

## Group By City


```python
print(df.groupby("City")["Sales"].sum())
```

## Group By Product


```python
print(df.groupby("Product")["Sales"].mean())
```

## Multiple Aggregations


```python
print(df.groupby("City")["Sales"].agg(["sum", "mean", "count"]))
```

## Group By Multiple Columns


```python
print(df.groupby(["City", "Product"])["Sales"].sum())
```

## Resetting Index


```python
grouped = df.groupby("City")["Sales"].sum().reset_index()
print(grouped)
```

## Why GroupBy is Important?


- Analyze sales by city
- Find best performing product
- Generate business insights
- Feature engineering for ML
