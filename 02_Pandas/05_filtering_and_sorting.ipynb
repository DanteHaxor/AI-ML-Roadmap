# Filtering and Sorting Data in Pandas 🔍

Filtering and sorting allow us to extract useful information
from large datasets, which is essential for Machine Learning.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
```

## Sample Dataset


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman", "Neha", "Riya"],
    "Age": [20, 21, 22, 19, 23],
    "Marks": [85, 90, 78, 92, 88]
}

df = pd.DataFrame(data)
print(df)
```

## Filtering Rows


```python
# Students with marks greater than 85
print(df[df["Marks"] > 85])
```

## Multiple Conditions


```python
print(df[(df["Marks"] > 85) & (df["Age"] > 20)])
```

## Using OR Condition


```python
print(df[(df["Marks"] > 90) | (df["Age"] < 20)])
```

## Sorting Data


```python
print(df.sort_values("Marks"))
print(df.sort_values("Marks", ascending=False))
```

## Sorting by Multiple Columns


```python
print(df.sort_values(["Age", "Marks"], ascending=[True, False]))
```

## Selecting Specific Columns


```python
print(df[["Name", "Marks"]])
```

## Why Filtering is Important in ML?


- Select training data
- Remove outliers
- Choose specific features
- Analyze patterns
