# Data Collection using APIs 🌐

APIs allow us to collect live, real-world data from websites
like weather, stocks, crypto, news, etc.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import requests
import pandas as pd
```

## What is an API?

An API (Application Programming Interface) allows programs
to communicate with websites and services.


## Example: Public API (JSONPlaceholder)


```python
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
data = response.json()

print(data[0])
```

## Convert API Data into Pandas DataFrame


```python
df = pd.DataFrame(data)
print(df.head())
```

## Selecting Useful Columns


```python
df = df[["id", "name", "email"]]
print(df)
```

## Saving API Data to CSV


```python
df.to_csv("users.csv", index=False)
print("users.csv saved")
```

## Why APIs Matter in Data Science?


- Collect live data
- Build datasets
- Train ML models on real information
