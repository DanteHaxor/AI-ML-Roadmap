# Web Scraping with Python 🌐

Web scraping is the process of extracting data from websites.
It is used when APIs are not available.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## What is Web Scraping?

Web scraping means:
- Downloading a web page
- Extracting useful data from HTML
- Converting it into structured data


## Example Website

We will scrape data from:
https://quotes.toscrape.com  
(A free website made for scraping practice)


```python
url = "https://quotes.toscrape.com"
response = requests.get(url)

print("Status Code:", response.status_code)
```

## Parse HTML using BeautifulSoup


```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.title.text)
```

## Extract Quotes from Website


```python
quotes = soup.find_all("span", class_="text")

for q in quotes[:5]:
    print(q.text)
```

## Extract Authors


```python
authors = soup.find_all("small", class_="author")

for a in authors[:5]:
    print(a.text)
```

## Store Data in List


```python
data = []

for q, a in zip(quotes, authors):
    data.append({
        "Quote": q.text,
        "Author": a.text
    })

print(data[:3])
```

## Convert to Pandas DataFrame


```python
df = pd.DataFrame(data)
print(df.head())
```

## Save Scraped Data to CSV


```python
df.to_csv("quotes.csv", index=False)
print("quotes.csv saved")
```

## Why Web Scraping Matters in Data Science?


- Collect data from websites
- Build custom datasets
- Useful when APIs are not available
