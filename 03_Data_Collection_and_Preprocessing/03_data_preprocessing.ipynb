# Data Preprocessing 🧹

Raw data cannot be used directly in Machine Learning.
Data preprocessing converts messy data into clean, usable form.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
import numpy as np
```

## Sample Raw Dataset


```python
data = {
    "Name": ["Abhay", "Rahul", "Aman", "Neha", "Riya"],
    "Age": [20, np.nan, 22, 19, np.nan],
    "Gender": ["M", "M", "M", "F", "F"],
    "Salary": [50000, 60000, np.nan, 55000, 52000]
}

df = pd.DataFrame(data)
print(df)
```

## Handling Missing Values


```python
print(df.isnull().sum())
```

## Fill Missing Values


```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
print(df)
```

## Encoding Categorical Data


```python
df["Gender"] = df["Gender"].map({"M": 0, "F": 1})
print(df)
```

## Feature Scaling


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[["Age", "Salary"]] = scaler.fit_transform(df[["Age", "Salary"]])

print(df)
```

## Splitting Features and Labels


```python
X = df.drop("Name", axis=1)
y = df["Name"]

print(X)
print(y)
```

## Why Preprocessing is Important?


- ML models need numerical input
- Missing values cause crashes
- Scaling improves accuracy
