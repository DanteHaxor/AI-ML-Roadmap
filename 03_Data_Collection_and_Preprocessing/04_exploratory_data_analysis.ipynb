# Exploratory Data Analysis (EDA) 📊

EDA is the process of understanding data using statistics and visualizations
before applying Machine Learning.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Sample Dataset


```python
data = {
    "Age": [20, 22, 19, 21, 23, 24, 22, 20],
    "Salary": [40000, 45000, 38000, 42000, 48000, 50000, 46000, 41000],
    "Gender": ["M", "F", "F", "M", "M", "F", "M", "F"]
}

df = pd.DataFrame(data)
print(df)
```

## Dataset Information


```python
print(df.info())
print(df.describe())
```

## Check Missing Values


```python
print(df.isnull().sum())
```

## Distribution of Age


```python
sns.histplot(df["Age"], kde=True)
plt.show()
```

## Salary Distribution


```python
sns.boxplot(x=df["Salary"])
plt.show()
```

## Gender Count


```python
sns.countplot(x="Gender", data=df)
plt.show()
```

## Relationship Between Age and Salary


```python
sns.scatterplot(x="Age", y="Salary", data=df)
plt.show()
```

## Correlation Matrix


```python
sns.heatmap(df.select_dtypes(include=["number"]).corr(), annot=True)
plt.show()
```

## Why EDA is Important?


- Find trends
- Detect outliers
- Understand relationships
- Improve ML model selection
