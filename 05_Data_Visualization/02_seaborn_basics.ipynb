# Data Visualization with Seaborn 🎨

Seaborn is built on Matplotlib and makes beautiful statistical graphs easily.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Sample Dataset


```python
data = {
    "Age": [20, 22, 19, 21, 23, 24, 22, 20],
    "Salary": [40000, 45000, 38000, 42000, 48000, 50000, 46000, 41000],
    "Gender": ["M", "F", "F", "M", "M", "F", "M", "F"]
}

df = pd.DataFrame(data)
print(df)
```

## Histogram with Seaborn


```python
sns.histplot(df["Age"], kde=True)
plt.show()
```

## Box Plot


```python
sns.boxplot(x="Gender", y="Salary", data=df)
plt.show()
```

## Count Plot


```python
sns.countplot(x="Gender", data=df)
plt.show()
```

## Scatter Plot


```python
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=df)
plt.show()
```

## Heatmap


```python
df["Gender_Num"] = df["Gender"].map({"M": 0, "F": 1})
sns.heatmap(df[["Age", "Salary", "Gender_Num"]].corr(), annot=True)
plt.show()
```
