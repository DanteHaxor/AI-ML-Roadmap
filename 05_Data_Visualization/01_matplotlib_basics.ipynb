# Data Visualization with Matplotlib 📊

Matplotlib is the foundation of all Python visualization libraries.
It is used to create plots for understanding data patterns.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Line Plot


```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Bar Chart


```python
names = ["Abhay", "Rahul", "Aman", "Neha"]
scores = [85, 90, 78, 92]

plt.bar(names, scores)
plt.title("Student Scores")
plt.show()
```

## Scatter Plot


```python
ages = [20, 22, 19, 21, 23]
salary = [40000, 45000, 38000, 42000, 48000]

plt.scatter(ages, salary)
plt.title("Age vs Salary")
plt.show()
```

## Histogram


```python
data = [10, 20, 20, 30, 30, 30, 40, 50]

plt.hist(data)
plt.title("Data Distribution")
plt.show()
```

## Why Visualization Matters?


- Understand patterns  
- Detect outliers  
- Present insights  
- Improve ML decisions
