# Statistical Visualization 📊

Statistical plots help us understand data distribution,
variability, and relationships.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Generate Random Dataset


```python
np.random.seed(42)

data = pd.DataFrame({
    "Height": np.random.normal(170, 10, 100),
    "Weight": np.random.normal(65, 8, 100),
    "Age": np.random.randint(18, 60, 100)
})

print(data.head())
```

## Distribution Plot


```python
sns.histplot(data["Height"], kde=True)
plt.title("Height Distribution")
plt.show()
```

## Box Plot


```python
sns.boxplot(x=data["Weight"])
plt.title("Weight Spread")
plt.show()
```

## Violin Plot


```python
sns.violinplot(x=data["Age"])
plt.title("Age Distribution")
plt.show()
```

## Pair Plot


```python
sns.pairplot(data)
plt.show()
```

## Correlation Heatmap


```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

## Why Statistical Visualization Matters?


- Understand distribution  
- Find outliers  
- Detect relationships  
- Improve ML model choice
