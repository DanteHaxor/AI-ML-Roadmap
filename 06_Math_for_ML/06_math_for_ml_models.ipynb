```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear Regression (Algebra + Statistics)


```python
# Simple dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Calculate slope and intercept using formula
m = np.cov(x, y)[0, 1] / np.var(x)
b = np.mean(y) - m * np.mean(x)

print("Slope:", m)
print("Intercept:", b)
```

## Prediction Line


```python
y_pred = m * x + b

plt.scatter(x, y)
plt.plot(x, y_pred)
plt.title("Linear Regression")
plt.show()
```

## Loss Function


```python
loss = np.mean((y - y_pred)**2)
print("Mean Squared Error:", loss)
```

## Gradient Descent


```python
m, b = 0, 0
lr = 0.01

for i in range(1000):
    y_pred = m * x + b
    dm = -2 * np.mean(x * (y - y_pred))
    db = -2 * np.mean(y - y_pred)

    m -= lr * dm
    b -= lr * db

print("Trained m:", m)
print("Trained b:", b)
```

## Why Math Matters?


- Loss functions come from statistics  
- Weights use algebra  
- Training uses calculus
