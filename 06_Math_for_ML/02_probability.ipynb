## Probability using Simulation


```python
coin_flips = np.random.choice(["H", "T"], size=1000)
prob_heads = np.sum(coin_flips == "H") / 1000

print("Estimated P(Heads):", prob_heads)
```

## Bayes’ Theorem 🧠

Bayes’ Theorem helps us update probabilities when new evidence is available.

Formula:

P(A | B) = (P(B | A) × P(A)) / P(B)

Where:
- P(A | B) = Probability of A given B  
- P(B | A) = Probability of B given A  
- P(A) = Probability of A  
- P(B) = Probability of B  


```python
# Example:
# Suppose:
# P(Disease) = 0.01
# P(Positive test | Disease) = 0.9
# P(Positive test) = 0.05

P_disease = 0.01
P_positive_given_disease = 0.9
P_positive = 0.05

P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

print("Probability of disease given positive test:", P_disease_given_positive)
```

## Why Bayes’ Theorem is Important in ML?

- Spam detection  
- Medical diagnosis  
- Recommendation systems  
- Naive Bayes classifiers
