# Statistics for Machine Learning 📊

Statistics helps us summarize, analyze, and understand data.
Machine Learning models rely heavily on statistical concepts.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import numpy as np
import pandas as pd
```

## Sample Dataset


```python
data = np.array([10, 20, 30, 40, 50, 60, 70, 80])
print(data)
```

## Mean (Average)


```python
print("Mean:", np.mean(data))
```

## Median


```python
print("Median:", np.median(data))
```

## Mode


```python
from scipy import stats

data2 = [10, 20, 20, 30, 40]
print("Mode:", stats.mode(data2))
```

## Variance


```python
print("Variance:", np.var(data))
```

## Standard Deviation


```python
print("Standard Deviation:", np.std(data))
```

## DataFrame Statistics


```python
df = pd.DataFrame({
    "Marks": [85, 90, 78, 92, 88],
    "Age": [20, 21, 22, 19, 23]
})

print(df.describe())
```

## Why Statistics is Important in ML?


- Understand data distribution  
- Detect outliers  
- Measure model performance
