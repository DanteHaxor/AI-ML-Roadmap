## Types of Numbers


```python
natural = [1, 2, 3, 4]
integers = [-2, -1, 0, 1, 2]
real_numbers = [1.5, 2.7, -3.2]

print(natural)
print(integers)
print(real_numbers)
```

## Sets


```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
```

## Why Sets Matter in ML?
