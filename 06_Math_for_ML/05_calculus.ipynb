# Calculus for Machine Learning 📈

Calculus is used to train ML models.
It helps minimize errors and find the best parameters.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Functions


```python
def f(x):
    return x**2

x = np.linspace(-10, 10, 100)
y = f(x)

plt.plot(x, y)
plt.title("y = x²")
plt.show()
```

## Derivative


```python
# Numerical derivative
dy_dx = np.gradient(y, x)

plt.plot(x, dy_dx)
plt.title("Derivative of y = x²")
plt.show()
```

## Gradient Descent Idea


```python
x = 5
learning_rate = 0.1

for i in range(10):
    gradient = 2 * x
    x = x - learning_rate * gradient
    print("Step", i, "x =", x)
```

## Why Calculus is Important?


- Trains ML models  
- Minimizes error  
- Optimizes weights
