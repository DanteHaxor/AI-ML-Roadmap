# SQL Joins 🔗

Joins are used to combine data from multiple tables.
Almost all real-world databases use joins.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("school.db")
cursor = conn.cursor()
```

## Create Tables


```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS marks (
    student_id INTEGER,
    subject TEXT,
    score INTEGER
)
""")

conn.commit()
```

## Insert Data


```python
cursor.execute("INSERT INTO students VALUES (1, 'Abhay')")
cursor.execute("INSERT INTO students VALUES (2, 'Rahul')")
cursor.execute("INSERT INTO students VALUES (3, 'Aman')")

cursor.execute("INSERT INTO marks VALUES (1, 'Math', 85)")
cursor.execute("INSERT INTO marks VALUES (1, 'Science', 90)")
cursor.execute("INSERT INTO marks VALUES (2, 'Math', 88)")
cursor.execute("INSERT INTO marks VALUES (3, 'Science', 75)")

conn.commit()
```

## Inner Join


```python
query = """
SELECT students.name, marks.subject, marks.score
FROM students
INNER JOIN marks
ON students.student_id = marks.student_id
"""

df = pd.read_sql(query, conn)
print(df)
```

## Left Join


```python
query = """
SELECT students.name, marks.subject, marks.score
FROM students
LEFT JOIN marks
ON students.student_id = marks.student_id
"""

df = pd.read_sql(query, conn)
print(df)
```

## Why Joins are Important?


- Combine data from multiple tables  
- Build complete datasets  
- Essential for ML pipelines
