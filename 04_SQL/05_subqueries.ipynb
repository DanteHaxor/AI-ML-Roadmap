# SQL Subqueries 🧠

A subquery is a query inside another query.
It helps solve complex problems using SQL.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("students.db")
```

## View Students Data


```python
df = pd.read_sql("SELECT * FROM students", conn)
print(df)
```

## Find Students Above Average Marks


```python
query = """
SELECT *
FROM students
WHERE marks > (
    SELECT AVG(marks) FROM students
)
"""

df = pd.read_sql(query, conn)
print(df)
```

## Subquery with SELECT


```python
query = """
SELECT name, marks,
       (SELECT AVG(marks) FROM students) AS avg_marks
FROM students
"""

df = pd.read_sql(query, conn)
print(df)
```

## Subquery with IN


```python
query = """
SELECT *
FROM students
WHERE id IN (
    SELECT id
    FROM students
    WHERE marks > 85
)
"""

df = pd.read_sql(query, conn)
print(df)
```

## Why Subqueries are Important?


- Solve complex queries  
- Create dynamic filters  
- Powerful data analysis
