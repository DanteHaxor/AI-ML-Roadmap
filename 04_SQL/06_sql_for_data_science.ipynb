```python
import sqlite3
import pandas as pd
```

## Create Sales Database


```python
conn = sqlite3.connect("sales.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    order_id INTEGER,
    product TEXT,
    city TEXT,
    amount INTEGER
)
""")

conn.commit()
```

## Insert Sales Data


```python
data = [
    (1, "Laptop", "Delhi", 50000),
    (2, "Phone", "Delhi", 30000),
    (3, "Laptop", "Mumbai", 55000),
    (4, "Phone", "Mumbai", 28000),
    (5, "Laptop", "Pune", 48000),
    (6, "Phone", "Pune", 25000)
]

cursor.executemany("INSERT INTO sales VALUES (?,?,?,?)", data)
conn.commit()
```

## Extract Data Using SQL


```python
query = """
SELECT city, product, SUM(amount) AS total_sales
FROM sales
GROUP BY city, product
"""

df = pd.read_sql(query, conn)
print(df)
```

## Convert SQL Data into Pandas


```python
pivot = df.pivot(index="city", columns="product", values="total_sales")
print(pivot)
```

## This Data is Now ML Ready


```python
print(pivot.fillna(0))
```

## Why SQL + Pandas is Powerful?


- SQL fetches the data  
- Pandas processes it  
- ML models train on it
