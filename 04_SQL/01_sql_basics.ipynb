# SQL Basics 🗄️

SQL (Structured Query Language) is used to store, retrieve,
and analyze data in databases.

Data Scientists use SQL to fetch data before applying ML.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import sqlite3
import pandas as pd
```

## Create Database and Connect


```python
conn = sqlite3.connect("students.db")
cursor = conn.cursor()
```

## Create Table


```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks INTEGER
)
""")
conn.commit()
```

## Insert Data


```python
cursor.execute("INSERT INTO students (name, age, marks) VALUES ('Abhay', 20, 85)")
cursor.execute("INSERT INTO students (name, age, marks) VALUES ('Rahul', 21, 90)")
cursor.execute("INSERT INTO students (name, age, marks) VALUES ('Aman', 22, 78)")
conn.commit()
```

## Read Data using SQL


```python
df = pd.read_sql("SELECT * FROM students", conn)
print(df)
```

## Why SQL is Important for ML?


- Most company data is in databases
- SQL extracts training data
- Pandas + SQL is a powerful combo
