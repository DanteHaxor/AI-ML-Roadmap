# SQL SELECT and WHERE 🗃️

`SELECT` is used to fetch data from a table.  
`WHERE` is used to filter records.

These are the most commonly used SQL commands in Data Science.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("students.db")
```

## View Full Table


```python
df = pd.read_sql("SELECT * FROM students", conn)
print(df)
```

## Select Specific Columns


```python
df = pd.read_sql("SELECT name, marks FROM students", conn)
print(df)
```

## Using WHERE Clause


```python
df = pd.read_sql("SELECT * FROM students WHERE marks > 80", conn)
print(df)
```

## Multiple Conditions


```python
df = pd.read_sql("SELECT * FROM students WHERE marks > 80 AND age > 20", conn)
print(df)
```

## Using OR Condition


```python
df = pd.read_sql("SELECT * FROM students WHERE marks > 85 OR age < 21", conn)
print(df)
```

## Using LIKE (Pattern Matching)


```python
df = pd.read_sql("SELECT * FROM students WHERE name LIKE 'A%'", conn)
print(df)
```

## Why SELECT & WHERE are Important?


- Fetch only required data  
- Filter huge databases  
- Prepare ML training data
