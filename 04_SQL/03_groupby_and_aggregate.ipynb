# SQL GROUP BY and Aggregate Functions 📊

`GROUP BY` is used to group rows that have the same values  
and apply functions like SUM, AVG, COUNT, etc.

This is extremely important in Data Analysis & Machine Learning.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("students.db")
```

## View Data


```python
df = pd.read_sql("SELECT * FROM students", conn)
print(df)
```

## Average Marks


```python
df = pd.read_sql("SELECT AVG(marks) AS average_marks FROM students", conn)
print(df)
```

## Count of Students


```python
df = pd.read_sql("SELECT COUNT(*) AS total_students FROM students", conn)
print(df)
```

## Group By Age


```python
df = pd.read_sql("""
SELECT age, AVG(marks) AS avg_marks
FROM students
GROUP BY age
""", conn)

print(df)
```

## Group By with COUNT


```python
df = pd.read_sql("""
SELECT age, COUNT(*) AS num_students
FROM students
GROUP BY age
""", conn)

print(df)
```

## Group By with Multiple Aggregates


```python
df = pd.read_sql("""
SELECT age,
       COUNT(*) AS total,
       AVG(marks) AS average,
       MAX(marks) AS max_marks,
       MIN(marks) AS min_marks
FROM students
GROUP BY age
""", conn)

print(df)
```

## Why GROUP BY is Important?


- Analyze performance by groups  
- Generate business insights  
- Create ML features
