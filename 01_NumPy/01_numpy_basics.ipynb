# NumPy Basics 🔢

NumPy (Numerical Python) is the core library for numerical computing in Python.
It is used heavily in Data Science, Machine Learning, and AI.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## Why NumPy?

Python lists are slow for numerical computation.
NumPy provides:
- Faster execution
- Less memory usage
- Mathematical functions
- Multi-dimensional arrays


```python
import numpy as np
```

## Creating NumPy Arrays


```python
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

arr2 = np.array([10, 20, 30])
print(arr2)
```

## Array Data Types


```python
arr = np.array([1, 2, 3.5])
print(arr)
print(arr.dtype)
```

## NumPy vs Python List


```python
py_list = [1, 2, 3, 4]
np_array = np.array([1, 2, 3, 4])

print(type(py_list))
print(type(np_array))
```

## NumPy Array Attributes


```python
arr = np.array([[1,2,3],[4,5,6]])

print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Size:", arr.size)
print("Data Type:", arr.dtype)
```

## Common Mistakes


```python
# This creates a Python list, not NumPy array
wrong = [1,2,3]

# This creates a NumPy array
correct = np.array([1,2,3])
```
