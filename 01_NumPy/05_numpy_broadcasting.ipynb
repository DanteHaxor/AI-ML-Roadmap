# NumPy Broadcasting 📡

Broadcasting allows NumPy to perform operations on arrays
of different shapes without making copies of data.

This concept is extremely important in Machine Learning.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import numpy as np
```

## What is Broadcasting?

Broadcasting is the ability of NumPy to stretch smaller arrays
to match larger arrays during arithmetic operations.


```python
arr = np.array([1, 2, 3])
print(arr + 10)
```

## Broadcasting with 1D and 2D Arrays


```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

vector = np.array([10, 20, 30])

print(matrix + vector)
```

## Broadcasting Rules


1. Compare shapes from right to left  
2. Dimensions must be equal or one must be 1  
3. If compatible, broadcasting happens


## Broadcasting Example (Row-wise)


```python
arr = np.array([[1,2,3],
                [4,5,6]])

row = np.array([10,20,30])

print(arr + row)
```

## Broadcasting Column-wise


```python
col = np.array([[10],
                [20]])

print(arr + col)
```

## Real ML Use Case


Broadcasting is used when:
- Adding bias to neurons
- Normalizing features
- Scaling datasets


## Common Broadcasting Error


```python
# This will cause an error
# a = np.array([1,2,3])
# b = np.array([1,2])
# print(a + b)
```
