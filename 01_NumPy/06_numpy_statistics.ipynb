# NumPy Statistics 📊

NumPy provides powerful statistical functions
to analyze and summarize datasets.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import numpy as np
```

## Sample Dataset


```python
data = np.array([10, 20, 30, 40, 50, 60, 70, 80])
print(data)
```

## Mean (Average)


```python
print("Mean:", np.mean(data))
```

## Median


```python
print("Median:", np.median(data))
```

## Variance


```python
print("Variance:", np.var(data))
```

## Standard Deviation


```python
print("Standard Deviation:", np.std(data))
```

## Minimum and Maximum


```python
print("Min:", np.min(data))
print("Max:", np.max(data))
```

## Percentiles


```python
print("25th percentile:", np.percentile(data, 25))
print("50th percentile:", np.percentile(data, 50))
print("75th percentile:", np.percentile(data, 75))
```

## Statistics on 2D Arrays


```python
arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("Column wise mean:", np.mean(arr, axis=0))
print("Row wise mean:", np.mean(arr, axis=1))
```

## Why These Stats Matter in ML?


- Feature normalization
- Outlier detection
- Data distribution analysis
- Model evaluation
