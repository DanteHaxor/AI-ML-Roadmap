```python
import numpy as np
```

## Sample Dataset (Student Marks)

Each row represents a student  
Columns: Math, Science, English


```python
marks = np.array([
    [80, 75, 90],
    [60, 70, 65],
    [85, 95, 88],
    [50, 60, 55],
    [90, 92, 85]
])

print(marks)
```

## Average Marks of Each Student


```python
student_avg = np.mean(marks, axis=1)
print("Student Averages:", student_avg)
```

## Average Marks of Each Subject


```python
subject_avg = np.mean(marks, axis=0)
print("Subject Averages:", subject_avg)
```

## Find Top Performer


```python
top_student = np.argmax(student_avg)
print("Top student index:", top_student)
print("Top student average:", student_avg[top_student])
```

## Students Who Passed (Average ≥ 70)


```python
passed = student_avg >= 70
print("Passed:", passed)
```

## Filtering Passed Students


```python
passed_students = marks[passed]
print(passed_students)
```

## Normalizing Data (Feature Scaling)


```python
min_vals = np.min(marks, axis=0)
max_vals = np.max(marks, axis=0)

normalized = (marks - min_vals) / (max_vals - min_vals)
print(normalized)
```

## Why This Matters in Machine Learning?


- Feature scaling improves model performance  
- Filtering is used to clean datasets  
- Statistics help detect patterns  
- NumPy arrays feed ML models
