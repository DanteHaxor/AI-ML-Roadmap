# NumPy Operations ⚙️

NumPy allows us to perform fast mathematical and matrix operations.
These operations are heavily used in Machine Learning.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import numpy as np
```

## Arithmetic Operations


```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

## Scalar Operations


```python
arr = np.array([1, 2, 3, 4])

print(arr + 10)
print(arr * 2)
```

## Mathematical Functions


```python
arr = np.array([1, 4, 9, 16])

print(np.sqrt(arr))
print(np.log(arr))
print(np.exp(arr))
```

## Aggregate Functions


```python
arr = np.array([10, 20, 30, 40])

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
print("Std:", np.std(arr))
```

## Matrix Operations


```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix Addition:\n", A + B)
print("Matrix Multiplication:\n", np.dot(A, B))
```

## Transpose of a Matrix


```python
print(A.T)
```

## Comparison Operations


```python
arr = np.array([10, 20, 30, 40])

print(arr > 25)
print(arr[arr > 25])
```

## Why These Operations Matter in ML?


- Feature scaling
- Matrix multiplication for neural networks
- Statistical calculations
- Data normalization
