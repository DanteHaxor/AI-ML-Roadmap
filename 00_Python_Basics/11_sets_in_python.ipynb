# Sets in Python 🧩

Sets are used to store multiple items in a single variable.
They are useful when working with **unique values**.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is a Set?

A set:
- Stores unique elements
- Is unordered
- Does not allow duplicates
- Is mutable


```python
numbers = {1, 2, 3, 4, 4, 5}
print(numbers)
```

## Empty Set

Use `set()` to create an empty set.


```python
empty_set = set()
print(type(empty_set))
```

## Accessing Set Elements

Sets are unordered, so elements cannot be accessed by index.


```python
fruits = {"apple", "banana", "cherry"}

for fruit in fruits:
    print(fruit)
```

## Adding Elements to a Set


```python
colors = {"red", "blue"}

colors.add("green")
print(colors)
```

## Removing Elements from Set


```python
colors.remove("red")
print(colors)

colors.discard("yellow")  # No error if element not present
```

## Common Set Methods
| Method      | Description           |
| ----------- | --------------------- |
| `add()`     | Add element           |
| `remove()`  | Remove element        |
| `discard()` | Remove without error  |
| `pop()`     | Remove random element |
| `clear()`   | Remove all elements   |


## Set Operations


```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

## Set Comprehension


```python
squares = {x**2 for x in range(1, 6)}
print(squares)
```

## Why Use Sets?

- Remove duplicates from lists
- Perform fast membership tests
- Mathematical set operations


## ⚠️ Common Mistakes

- Creating empty set using `{}` (creates dictionary)
- Expecting ordered output
- Using mutable elements inside sets
