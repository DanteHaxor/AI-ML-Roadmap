# Constructors and Methods in Python OOPS 🏗️

Constructors and methods help initialize objects and define their behavior.
They are core concepts of Object-Oriented Programming in Python.

## What is a Constructor?

A constructor is a **special method** that is automatically called
when an object is created.

In Python, a constructor is defined using:
init()

```python
class Student:
    def __init__(self):
        print("Student object created")

s1 = Student()
```

## Constructor with Parameters

Constructors can accept parameters to initialize object attributes.


```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Abhay", 20)
s2 = Student("Rahul", 21)

print(s1.name, s1.age)
print(s2.name, s2.age)
```

## Instance Attributes

Instance attributes belong to an object.
Each object has its own copy.


```python
class Car:
    def __init__(self, brand):
        self.brand = brand

c1 = Car("Tesla")
c2 = Car("BMW")

print(c1.brand)
print(c2.brand)
```

## Class Attributes

Class attributes are shared among all objects of the class.


```python
class Car:
    wheels = 4  # class attribute

    def __init__(self, brand):
        self.brand = brand

c1 = Car("Tesla")
c2 = Car("BMW")

print(c1.wheels)
print(c2.wheels)
```

## Instance Methods

Instance methods work with instance attributes and use `self`.


```python
class Student:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello", self.name)

s1 = Student("Abhay")
s1.greet()
```

## Class Methods

Class methods work with class attributes.
They use `cls` and `@classmethod`.


```python
class Car:
    wheels = 4

    @classmethod
    def show_wheels(cls):
        print("Wheels:", cls.wheels)

Car.show_wheels()
```

## Static Methods

Static methods do not use `self` or `cls`.
They behave like normal functions inside a class.


```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.add(10, 20))
```

## Instance vs Class vs Static Methods

| Method Type | Uses |
|-----------|------|
| Instance | Object data |
| Class | Class-level data |
| Static | Utility logic |
