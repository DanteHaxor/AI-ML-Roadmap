# Loops in Python 🔁

Loops are used to execute a block of code repeatedly.
They are essential for automation, data processing, and AI/ML workflows.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What are Loops?

Loops allow us to run the same code multiple times until a condition is met.

Python provides:
- `while` loop
- `for` loop


## while Loop

The `while` loop runs as long as the given condition is **True**.


```python
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
```

## ⚠️ Infinite Loop

If the condition never becomes False, the loop runs forever.
Always update the loop variable.


## for Loop

The `for` loop is used to iterate over a sequence such as
a list, tuple, string, or range.


```python
for i in range(1, 6):
    print(i)
```

## range() Function

`range(start, stop, step)`
- start → inclusive
- stop → exclusive
- step → optional


```python
print(list(range(5)))
print(list(range(1, 10)))
print(list(range(1, 10, 2)))
```

## break Statement

The `break` statement is used to stop the loop immediately.


```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

## continue Statement

The `continue` statement skips the current iteration and moves to the next one.


```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

## Loops with User Input


```python
n = int(input("Enter a number: "))

for i in range(1, n + 1):
    print(i)
```

## Nested Loops

A loop inside another loop is called a nested loop.


```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

## ⚠️ Common Mistakes

- Forgetting to update loop variables
- Off-by-one errors in range()
- Misusing break and continue
