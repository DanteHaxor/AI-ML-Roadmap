# Exception Handling in Python ⚠️

Exception handling allows a program to handle runtime errors
gracefully without crashing.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is an Exception?

An exception is an **error that occurs during program execution**.

Common exceptions:
- ZeroDivisionError
- ValueError
- TypeError
- FileNotFoundError


## Why Handle Exceptions?

- Prevent program crashes
- Show meaningful error messages
- Maintain smooth program flow


```python
# This will cause an error
# result = 10 / 0
```

## try-except Block

Used to catch and handle exceptions.


```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

## Handling Multiple Exceptions


```python
try:
    num = int("abc")
except ZeroDivisionError:
    print("Division error")
except ValueError:
    print("Value conversion error")
```

## Generic Exception Handling


```python
try:
    x = int(input("Enter a number: "))
    print(10 / x)
except Exception as e:
    print("Error:", e)
```

## else Block

The `else` block runs if no exception occurs.


```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print(a / b)
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful")
```

## finally Block

The `finally` block always executes,
whether an exception occurs or not.


```python
try:
    file = open("test.txt", "r")
except FileNotFoundError:
    print("File not found")
finally:
    print("Execution completed")
```

## Raising Exceptions

Use `raise` to manually trigger exceptions.


```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

## ⚠️ Common Mistakes

- Catching exceptions too broadly
- Ignoring exceptions silently
- Overusing try-except blocks
