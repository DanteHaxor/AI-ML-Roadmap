# File Handling in Python 📁

File handling allows a program to read from and write data to files.
It is very important for logging, data storage, and data processing.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is File Handling?

File handling means:
- Creating files
- Reading data from files
- Writing data to files
- Updating and deleting files


## File Modes in Python

| Mode | Description |
|----|------------|
| `r` | Read (default) |
| `w` | Write (overwrite) |
| `a` | Append |
| `x` | Create |
| `rb` | Read binary |
| `wb` | Write binary |


```python
file = open("example.txt", "w")
file.write("Hello, this is a file handling example.")
file.close()
```

## Reading from a File


```python
file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
```

## Append Mode (`a`)

Append adds data to the end of the file.


```python
file = open("example.txt", "a")
file.write("\nThis line is appended.")
file.close()
```

## `with` Keyword

The `with` statement automatically closes the file.
It is the recommended way to work with files.


```python
with open("example.txt", "r") as file:
    print(file.read())
```

## Reading File Line by Line


```python
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

## Deleting Files

Python provides `os` module to delete files.


```python
import os

if os.path.exists("example.txt"):
    os.remove("example.txt")
    print("File deleted")
else:
    print("File does not exist")
```

## ⚠️ Common Mistakes

- Forgetting to close file
- Using wrong file mode
- Trying to read non-existing file
