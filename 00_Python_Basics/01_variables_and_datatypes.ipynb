## What is a Variable?

A variable is a **container** used to store data in memory.

In Python:
- You do not need to declare the data type
- The type is decided at runtime


```python
name = "Abhay"
age = 20
height = 5.8
is_student = True

print(name)
print(age)
print(height)
print(is_student)
```

## Rules for Naming Variables

✅ Valid:
- Must start with a letter (a-z, A-Z) or underscore `_`
- Can contain letters, numbers, and underscores
- Case-sensitive

❌ Invalid:
- Cannot start with a number
- Cannot use Python keywords


```python
# Valid variable names
user_name = "Rahul"
_age = 25

# Invalid (uncomment to see error)
# 1name = "Test"
```

## Data Types in Python

Python has several built-in data types:

| Type | Description | Example |
|----|----|----|
| int | Integer numbers | 10 |
| float | Decimal numbers | 3.14 |
| str | Text | "Hello" |
| bool | Boolean | True / False |


```python
x = 10
y = 3.5
z = "Python"
flag = True

print(type(x))
print(type(y))
print(type(z))
print(type(flag))
```

## Keywords and Comments

### Keywords
Keywords are **reserved words** in Python.
Examples: `if`, `else`, `for`, `while`, `True`, `False`, `None`

### Comments
Comments are used to explain code.
They are ignored by Python during execution.


```python
# This is a single-line comment

"""
This is a
multi-line comment
"""

x = 5  # Assigning value to x
print(x)
```
