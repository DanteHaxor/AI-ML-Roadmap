# Tuples in Python 📦

Tuples are used to store multiple items in a single variable.
They are similar to lists but **immutable**.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is a Tuple?

A tuple is:
- Ordered
- Immutable (cannot be changed)
- Allows duplicate values
- Can store different data types


```python
numbers = (1, 2, 3, 4)
names = ("Abhay", "Rahul", "Aman")
mixed = (1, "Python", 3.5, True)

print(numbers)
print(names)
print(mixed)
```

## Single Element Tuple

A tuple with one element **must have a comma**.


```python
single = (10,)
print(type(single))

not_tuple = (10)
print(type(not_tuple))
```

## Accessing Tuple Elements

Tuple indexing and slicing work similar to lists.


```python
t = (10, 20, 30, 40, 50)

print(t[0])
print(t[-1])
print(t[1:4])
```

## Tuple Immutability

Tuples cannot be modified after creation.


```python
t = (1, 2, 3)

# This will cause an error
# t[0] = 10
```

## Tuple Methods

Tuples have only two built-in methods:
| Method    | Description       |
| --------- | ----------------- |
| `count()` | Count occurrences |
| `index()` | Find index        |


```python
t = (1, 2, 3, 2, 4, 2)

print(t.count(2))
print(t.index(3))
```

## Tuple Packing and Unpacking


```python
# Packing
student = ("Abhay", 20, "AI/ML")

# Unpacking
name, age, course = student

print(name)
print(age)
print(course)
```

## Using Tuples with Loops


```python
colors = ("red", "green", "blue")

for color in colors:
    print(color)
```

## Tuple vs List
| Tuple       | List         |
| ----------- | ------------ |
| Immutable   | Mutable      |
| Faster      | Slower       |
| Less memory | More memory  |
| Fixed data  | Dynamic data |


## ⚠️ Common Mistakes

- Forgetting comma in single-element tuple
- Trying to modify tuple elements
