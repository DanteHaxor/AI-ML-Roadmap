# Object-Oriented Programming (OOPS) Basics in Python 🧠

Object-Oriented Programming (OOPS) is a programming paradigm
that uses **objects and classes** to design programs.

## Why Use OOPS?

OOPS helps to:
- Organize complex programs
- Reuse code efficiently
- Improve readability and scalability
- Model real-world entities

OOPS is widely used in large applications and AI/ML frameworks.


## Key Concepts of OOPS

- Class
- Object
- Attributes
- Methods
- Encapsulation
- Inheritance
- Polymorphism
- Abstraction


## What is a Class?

A class is a **blueprint** for creating objects.
It defines the properties (attributes) and behaviors (methods).


```python
class Student:
    name = "Abhay"
    age = 20
```

## What is an Object?

An object is an **instance** of a class.


```python
s1 = Student()

print(s1.name)
print(s1.age)
```

## Attributes

Attributes are variables that belong to a class or object.


```python
s1.name = "Rahul"
s1.age = 21

print(s1.name)
print(s1.age)
```

## Methods

Methods are functions defined inside a class.
They describe the behavior of objects.


```python
class Student:
    def greet(self):
        print("Hello, welcome to OOPS")

s1 = Student()
s1.greet()
```

## The `self` Keyword

`self` refers to the current object.
It is used to access instance variables and methods.


```python
class Student:
    def set_name(self, name):
        self.name = name

    def show_name(self):
        print(self.name)

s1 = Student()
s1.set_name("Abhay")
s1.show_name()
```

## Creating Multiple Objects from Same Class


```python
s1 = Student()
s2 = Student()

s1.set_name("Abhay")
s2.set_name("Rahul")

s1.show_name()
s2.show_name()
```

## Real-Life Example of Class & Object


```python
class Car:
    brand = "Tesla"

    def start(self):
        print("Car started")

c1 = Car()
print(c1.brand)
c1.start()
```

## ⚠️ Common Mistakes

- Forgetting `self` in method definition
- Accessing attributes without object
- Confusing class and object
