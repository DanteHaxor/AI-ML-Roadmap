# Conditional Statements in Python 🔀

Conditional statements allow a program to make decisions based on conditions.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What are Conditional Statements?

Conditional statements execute different blocks of code based on
whether a condition is **True** or **False**.

Python supports:
- `if`
- `if-else`
- `if-elif-else`


```python
age = 18

if age >= 18:
    print("You are eligible to vote")
```

## if-else Statement

Used when we want to execute one block if the condition is true,
and another block if it is false.


```python
num = 7

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

## if-elif-else Statement

Used when multiple conditions need to be checked.


```python
marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: D")
```

## Nested if Statements

An `if` statement inside another `if` statement is called nesting.


```python
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Entry allowed")
    else:
        print("ID required")
else:
    print("Underage")
```

## match-case Statement (Python 3.10+)

`match-case` is similar to switch-case in other languages.
It is used when we have multiple fixed choices.


```python
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case _:
        print("Invalid day")
```

## Conditional Statements with User Input


```python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

## ⚠️ Common Mistakes
