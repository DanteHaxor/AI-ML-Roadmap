## Why Type Conversion is Important?

- User input is always taken as a **string**
- Mathematical operations require numeric types
- Data Science often needs type conversions while cleaning data


## Types of Type Conversion

There are two types of type conversion in Python:

1. **Implicit Type Conversion**
2. **Explicit Type Conversion (Type Casting)**


```python
# Python automatically converts int to float
a = 10      # int
b = 3.5     # float

result = a + b
print(result)
print(type(result))
```

## Explicit Type Conversion (Type Casting)

We manually convert one data type into another using built-in functions:
- int()
- float()
- str()
- bool()


```python
x = "100"

# String to integer
num = int(x)
print(num, type(num))

# Integer to float
f = float(num)
print(f, type(f))

# Integer to string
s = str(num)
print(s, type(s))
```

## Boolean Type Conversion

Rules:
- `0`, `None`, `""` → False
- Non-zero numbers and non-empty strings → True


```python
print(bool(0))
print(bool(1))
print(bool(""))
print(bool("Python"))
```

## Taking User Input

Python uses the `input()` function to take input from the user.
⚠️ Input is always stored as a **string**.


```python
name = input("Enter your name: ")
print("Hello", name)
```

## User Input with Type Conversion

To perform calculations, convert input into numeric types.


```python
age = int(input("Enter your age: "))
print("Your age after 5 years will be:", age + 5)
```

## Taking Multiple Inputs from User


```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum:", a + b)
print("Product:", a * b)
```

## ⚠️ Common Errors

- `ValueError` when converting invalid strings to numbers
- Forgetting to convert input before calculation
