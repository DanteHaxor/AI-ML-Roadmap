# Lists in Python 📋

Lists are used to store multiple items in a single variable.
They are one of the most important data structures in Python.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is a List?

A list is:
- Ordered
- Mutable (changeable)
- Allows duplicate values
- Can store different data types


```python
numbers = [1, 2, 3, 4, 5]
names = ["Abhay", "Rahul", "Aman"]
mixed = [1, "Python", 3.5, True]

print(numbers)
print(names)
print(mixed)
```

## Accessing List Elements

List indexing starts from 0.
Negative indexing starts from -1.


```python
fruits = ["apple", "banana", "cherry"]

print(fruits[0])
print(fruits[1])
print(fruits[-1])
```

## List Slicing

Slicing extracts a portion of the list.

Syntax:
list[start : end : step]

```python
nums = [0, 1, 2, 3, 4, 5, 6]

print(nums[1:5])
print(nums[:4])
print(nums[3:])
print(nums[::2])
print(nums[::-1])
```

## Modifying Lists

Lists are mutable, so we can change their elements.


```python
colors = ["red", "blue", "green"]
colors[1] = "yellow"

print(colors)
```

## Common List Methods
| Method      | Description     |
| ----------- | --------------- |
| `append()`  | Add element     |
| `insert()`  | Insert at index |
| `remove()`  | Remove element  |
| `pop()`     | Remove by index |
| `sort()`    | Sort list       |
| `reverse()` | Reverse list    |
| `count()`   | Count element   |
| `index()`   | Find index      |


```python
nums = [3, 1, 4, 1, 5]

nums.append(9)
nums.insert(1, 7)
nums.remove(1)
nums.pop()
nums.sort()
nums.reverse()

print(nums)
```

## Using Loops with Lists


```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

## List Comprehensions

List comprehensions provide a concise way to create lists.


```python
squares = [x**2 for x in range(1, 6)]
print(squares)
```

## List Comprehension with Condition


```python
even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print(even_numbers)
```

## ⚠️ Common Mistakes

- IndexError due to invalid index
- Confusing `append()` and `extend()`
- Modifying list while iterating
