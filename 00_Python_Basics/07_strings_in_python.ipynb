# Strings in Python 🧵

Strings are used to store and manipulate text.
They are one of the most commonly used data types in Python.

## What is a String?

A string is a sequence of characters enclosed in:
- Single quotes `' '`
- Double quotes `" "`
- Triple quotes `''' '''` or `""" """`


```python
name = "Python"
course = 'AI/ML'
description = """This is a
multi-line string"""

print(name)
print(course)
print(description)
```

## Accessing Characters in a String

Strings are **indexed** starting from 0.
Negative indexing starts from -1.


```python
text = "Python"

print(text[0])
print(text[3])
print(text[-1])
```

## String Slicing

Slicing extracts a portion of a string.

Syntax:
string[start : end : step]


```python
word = "Programming"

print(word[0:5])
print(word[3:8])
print(word[:4])
print(word[4:])
print(word[::2])
print(word[::-1])  # Reverse string
```

## String Immutability

Strings are **immutable**, meaning they cannot be changed after creation.


```python
name = "Python"

# This will cause an error
# name[0] = 'J'
```

## Common String Methods
| Method      | Description          |
| ----------- | -------------------- |
| `upper()`   | Convert to uppercase |
| `lower()`   | Convert to lowercase |
| `strip()`   | Remove spaces        |
| `replace()` | Replace text         |
| `split()`   | Split string         |
| `find()`    | Find index           |


```python
msg = "  hello Python  "

print(msg.upper())
print(msg.lower())
print(msg.strip())
print(msg.replace("Python", "World"))
print(msg.split())
print(msg.find("Python"))
```

## String Formatting

String formatting is used to insert variables into strings.
Python supports:
- f-strings
- format() method


```python
name = "Abhay"
age = 20

print(f"My name is {name} and I am {age} years old.")
```

## format() Method


```python
name = "Abhay"
course = "AI/ML"

print("My name is {} and I am learning {}".format(name, course))
```

## Escape Characters

Escape characters are used to format strings.
Examples:
- `\n` → New line
- `\t` → Tab


```python
print("Hello\nWorld")
print("Python\tProgramming")
```
