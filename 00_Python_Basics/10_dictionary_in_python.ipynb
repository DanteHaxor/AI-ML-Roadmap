# Dictionary in Python 🗂️

Dictionaries are used to store data in **key-value pairs**.
They are very important in data handling and real-world applications.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is a Dictionary?

A dictionary:
- Stores data as **key : value**
- Is unordered (in older versions)
- Is mutable
- Does not allow duplicate keys


```python
student = {
    "name": "Abhay",
    "age": 20,
    "course": "AI/ML"
}

print(student)
```

## Accessing Dictionary Values

Values are accessed using keys.


```python
print(student["name"])
print(student.get("age"))
```

## Modifying Dictionary

Dictionaries are mutable.
You can add or update values.


```python
student["age"] = 21
student["city"] = "Delhi"

print(student)
```

## Removing Elements from Dictionary


```python
student.pop("city")
del student["course"]

print(student)
```

## Common Dictionary Methods
| Method     | Description         |
| ---------- | ------------------- |
| `keys()`   | Get all keys        |
| `values()` | Get all values      |
| `items()`  | Get key-value pairs |
| `update()` | Update dictionary   |
| `clear()`  | Remove all items    |


```python
student = {"name": "Abhay", "age": 21}

print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 22})
print(student)
```

## Looping Through a Dictionary


```python
for key, value in student.items():
    print(key, ":", value)
```

## Nested Dictionary

A dictionary inside another dictionary.


```python
students = {
    "student1": {"name": "Abhay", "age": 20},
    "student2": {"name": "Rahul", "age": 21}
}

print(students["student1"]["name"])
```

## Dictionary Comprehension

Create dictionaries using a concise syntax.


```python
squares = {x: x**2 for x in range(1, 6)}
print(squares)
```

## ⚠️ Common Mistakes

- Using mutable keys (like lists)
- Accessing non-existing keys directly
- Confusing `[]` and `get()`
