# 🧠 Linear Regression

Linear Regression is one of the most fundamental algorithms in Machine Learning.

It is used to predict **continuous values**.

---

## 💡 Idea

👉 Find a straight line that best fits the data

Equation of a line:

y = mx + b

- m → slope (weight)
- b → intercept (bias)

---

## 🎯 Goal

Find the best values of m and b such that:
👉 Error between predicted and actual values is minimum

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# 📊 Creating a Simple Dataset

We create a small dataset for visualization

```python
# Simple linear data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Dataset")
plt.show()
```

# 📐 Ordinary Least Squares (OLS)

OLS is a **mathematical method** to find best line.

---

## 💡 Idea

Minimize:

👉 Sum of squared errors

Error = (Actual - Predicted)²

---

## 🎯 Objective

Find m and b such that total error is minimum

This formula directly computes the best weights.

👉 No iteration needed  
👉 Exact solution (if matrix is invertible)

```python
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
plt.scatter(X, y, label="Actual")
plt.plot(X, y_pred, color='red', label="OLS Line")
plt.legend()
plt.title("OLS Fit")
plt.show()
```

# ⚠️ Limitations of OLS

- Requires matrix inversion
- Slow for large datasets
- Cannot scale well for millions of rows

---

## 💡 Solution

👉 Use Gradient Descent

# 🔄 Gradient Descent

Gradient Descent is an **iterative optimization algorithm**.

---

## 💡 Idea

1. Start with random m and b
2. Calculate error
3. Update values step by step
4. Repeat until error is minimized

---

## 🎯 Goal

Reach the minimum point of error function

Gradient Descent updates parameters using derivatives
θ=θ−α∇J(θ)

- θ → parameters (m, b)
- α → learning rate
- ∇J → gradient

---

## 💡 Intuition

👉 Move step-by-step towards minimum error

```python
# Initialize parameters
m = 0
b = 0
learning_rate = 0.01
epochs = 1000

n = len(X)

# Flatten X for simplicity
X_flat = X.flatten()

for _ in range(epochs):
    y_pred = m * X_flat + b
    
    # gradients
    dm = (-2/n) * np.sum(X_flat * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    # update
    m = m - learning_rate * dm
    b = b - learning_rate * db

print("GD Slope:", m)
print("GD Intercept:", b)
```

```python
y_pred_gd = m * X_flat + b

plt.scatter(X, y, label="Actual")
plt.plot(X, y_pred, color='red', label="OLS")
plt.plot(X, y_pred_gd, color='green', linestyle='--', label="GD")
plt.legend()
plt.title("OLS vs Gradient Descent")
plt.show()
```

# 📌 OLS vs Gradient Descent

| Feature            | OLS                     | Gradient Descent        |
|------------------|------------------------|------------------------|
| Approach         | Analytical             | Iterative              |
| Speed            | Fast (small data)      | Slower per iteration   |
| Scalability      | Poor for large data    | Scales well            |
| Accuracy         | Exact solution         | Approximate            |

---

## 💡 Insight

👉 OLS = exact but heavy  
👉 GD = flexible and scalable  

# 🎯 Final Takeaways

- Linear Regression fits a straight line
- OLS gives direct mathematical solution
- Gradient Descent learns step-by-step

---

## 🚀 When to Use?

✔ OLS → small datasets  
✔ GD → large datasets / deep learning  

---

## 💥 Golden Concept

👉 Almost all ML & Deep Learning uses Gradient Descent
