# Linear Regression – Mathematical Intuition 📐

Linear Regression is the foundation of supervised learning.
It models the relationship between input features and a continuous target.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is Linear Regression?

Linear Regression tries to find the **best-fit line** that explains
the relationship between input (X) and output (y).

The idea is to minimize prediction error.


## Equation of a Line


y = mx + b

Where:
- m = slope (weight)
- b = intercept (bias)
- x = input feature
- y = predicted output


## Best Fit Line – Intuition


The **best-fit line** is the line for which:
- Predictions are closest to actual values
- Overall error is minimized


## Visual Understanding


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.scatter(x, y)
plt.plot(x, y, label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Error (Residual)


Residual = Actual value − Predicted value

Residuals show how far predictions are from real values.


## Objective of Linear Regression


The goal is to find values of **m** and **b** that minimize
the total error across all data points.


## Assumptions of Linear Regression


- Linear relationship between X and y  
- Errors are normally distributed  
- No multicollinearity  
- Homoscedasticity (constant variance)  


## Why Learn Linear Regression First?


- Foundation of many ML algorithms  
- Easy to interpret  
- Builds intuition for optimization
