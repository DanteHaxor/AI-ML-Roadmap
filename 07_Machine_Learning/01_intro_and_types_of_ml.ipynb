# Introduction to Machine Learning 🤖

Machine Learning (ML) is a subset of Artificial Intelligence
that enables systems to learn patterns from data and make predictions
without being explicitly programmed.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is Machine Learning?

Machine Learning allows computers to:
- Learn from past data
- Identify patterns
- Make decisions or predictions

Instead of writing rules manually, we let data define the rules.


## Why Machine Learning is Important?

Machine Learning powers:
- Search engines
- Recommendation systems
- Spam filters
- Fraud detection
- Self-driving cars


## Types of Machine Learning


### 1️⃣ Supervised Learning

- Data is labeled
- Model learns input → output mapping

Examples:
- House price prediction
- Email spam detection


### 2️⃣ Unsupervised Learning

- Data is unlabeled
- Model finds hidden patterns

Examples:
- Customer segmentation
- Market basket analysis


### 3️⃣ Reinforcement Learning

- Agent learns by trial and error
- Feedback is given as rewards

Examples:
- Game playing AI
- Robotics


## Machine Learning Workflow


Data Collection → Preprocessing → EDA → Feature Engineering  
→ Model Training → Evaluation → Prediction
