# Supervised Learning Overview 📘

Supervised Learning is the most widely used type of Machine Learning.
In this approach, models learn from **labeled data**.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is Supervised Learning?

In supervised learning:
- Each data point has an **input (X)** and a **label/output (y)**
- The model learns a mapping from X → y
- The goal is to predict y for unseen X


## Example of Supervised Learning


| Input (Features) | Output (Label) |
|------------------|----------------|
| Size of house, Location | House price |
| Email text | Spam / Not Spam |
| Exam scores | Pass / Fail |


## Types of Supervised Learning


### 1️⃣ Regression

- Output is **continuous**
- Predicts numerical values

Examples:
- House price prediction
- Salary prediction
- Stock price estimation


### 2️⃣ Classification

- Output is **categorical**
- Predicts class labels

Examples:
- Spam detection
- Disease diagnosis
- Fraud detection


## Key Components of Supervised Learning


- **Features (X):** Input variables  
- **Target (y):** Output variable  
- **Model:** Learns patterns  
- **Loss Function:** Measures error  
- **Optimizer:** Minimizes loss  


## Training vs Testing Data


- Training data is used to learn patterns  
- Testing data is used to evaluate performance  
- Helps detect overfitting and underfitting  


## Why Supervised Learning is Important?


- Most real-world ML problems are supervised  
- Easy to evaluate performance  
- High accuracy when labeled data is available
