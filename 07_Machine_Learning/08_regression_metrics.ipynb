# Regression Evaluation Metrics 📏

After training a regression model, we must evaluate
how well it performs using proper metrics.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## Sample Actual vs Predicted Values


```python
y_actual = np.array([100, 150, 200, 250, 300])
y_pred = np.array([110, 140, 210, 240, 290])
```

## Mean Absolute Error (MAE)


```python
mae = mean_absolute_error(y_actual, y_pred)
print("MAE:", mae)
```

## Mean Squared Error (MSE)


```python
mse = mean_squared_error(y_actual, y_pred)
print("MSE:", mse)
```

## Root Mean Squared Error (RMSE)


```python
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

## R² Score (Coefficient of Determination)


```python
r2 = r2_score(y_actual, y_pred)
print("R² Score:", r2)
```

## Interpretation of Metrics


- **MAE**: Average absolute error  
- **MSE**: Penalizes large errors  
- **RMSE**: Error in original units  
- **R²**: How well model explains variance  


## Which Metric to Use?


- Use MAE for robustness  
- Use RMSE when large errors matter  
- Use R² for overall model fit  


## Why Metrics Matter?


- Compare models  
- Detect underfitting / overfitting  
- Choose best algorithm
