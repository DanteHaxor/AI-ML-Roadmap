# Cost Function and Gradient Descent 📉

To train a Linear Regression model, we need:
1. A way to measure error → Cost Function
2. A way to minimize error → Gradient Descent

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**


## What is a Cost Function?

A cost function measures how wrong our model predictions are.
Lower cost = better model.


## Mean Squared Error (MSE)

The most common cost function for Linear Regression:

MSE = (1/n) Σ (y_actual − y_predicted)²


```python
import numpy as np
```

```python
# Example data
y_actual = np.array([2, 4, 6, 8])
y_pred = np.array([1.8, 3.9, 6.2, 7.9])

mse = np.mean((y_actual - y_pred) ** 2)
print("Mean Squared Error:", mse)
```

## Cost Function Curve

The cost function creates a curve showing
how error changes with model parameters.


```python
import matplotlib.pyplot as plt

m_values = np.linspace(-1, 5, 100)
costs = []

x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

for m in m_values:
    y_pred = m * x
    costs.append(np.mean((y - y_pred) ** 2))

plt.plot(m_values, costs)
plt.xlabel("Slope (m)")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function Curve")
plt.show()
```

## What is Gradient Descent?

Gradient Descent is an optimization algorithm used to:
- Start with random parameters
- Move step-by-step toward minimum cost


## Update Rule

For Linear Regression:

m = m − α × ∂(cost)/∂m  
b = b − α × ∂(cost)/∂b  

Where α is the learning rate.


```python
# Gradient Descent Example
m, b = 0.0, 0.0
learning_rate = 0.01
epochs = 1000

x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

for _ in range(epochs):
    y_pred = m * x + b
    dm = -2 * np.mean(x * (y - y_pred))
    db = -2 * np.mean(y - y_pred)

    m -= learning_rate * dm
    b -= learning_rate * db

print("Optimized m:", m)
print("Optimized b:", b)
```

## Learning Rate (α)

- Too small → slow learning  
- Too large → overshoot minimum  
- Proper value → fast convergence  


## Why Gradient Descent is Important?

Almost every ML model:
- Minimizes a cost function
- Uses gradient-based optimization
