# Dummy Variables & Encoding 🧩

Most Machine Learning algorithms work only with numbers.
Categorical data must be converted into numerical form.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**

## What are Categorical Variables?

Categorical variables represent categories or labels instead of numbers.

Examples:
- Gender → Male, Female  
- City → Delhi, Mumbai, Pune  
- Purchased → Yes, No  

## Why Encoding is Required?

- ML models perform mathematical operations  
- Text data cannot be processed directly  
- Encoding converts categories → numbers  

## Example Dataset

```python
import pandas as pd

df = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Pune", "Delhi"],
    "Purchased": ["Yes", "No", "Yes", "No"],
    "Price": [50000, 60000, 45000, 52000]
})

print(df)
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Purchased_LE"] = le.fit_transform(df["Purchased"])
print(df)
```

## One-Hot Encoding (Dummy Variables)

```python
df_dummies = pd.get_dummies(df, columns=["City"], drop_first=True)
print(df_dummies)
```

## Dummy Variable Trap

The dummy variable trap occurs when:
- One dummy variable can be predicted from others
- Causes multicollinearity in linear models

Solution:
👉 Drop one dummy column (`drop_first=True`)

## Encoding using scikit-learn

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df[["City"]])

print(encoded)

## When to Use Which Encoding?

- Label Encoding → Ordinal data  
- One-Hot Encoding → Nominal data  
- Avoid label encoding for unordered categories
