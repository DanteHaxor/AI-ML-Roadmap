# Overfitting vs Underfitting ⚖️

When training machine learning models, two common problems occur:

- Underfitting
- Overfitting

Understanding these issues is critical for building good ML models.

📌 Part of: **AI/ML Roadmap – Beginner to Advanced**

## What is Underfitting?

Underfitting occurs when the model is **too simple**
to capture the underlying pattern of the data.

The model performs poorly on:
- Training data
- Test data

### Example of Underfitting

Imagine fitting a straight line to data that follows a curve.

The model fails to learn the true relationship.

## What is Overfitting?

Overfitting occurs when the model **learns noise instead of patterns**.

The model performs:
- Very well on training data
- Poorly on unseen data

### Example of Overfitting

A very complex model may memorize the training dataset
instead of learning general patterns.

## Visual Intuition

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)

plt.scatter(x, y)
plt.title("Data Example")
plt.show()
```

### Model Complexity vs Error

- Low complexity → Underfitting  
- Optimal complexity → Best model  
- High complexity → Overfitting  

## How to Detect Overfitting?

Compare training and validation performance.

If:
- Training error is low
- Validation error is high

➡️ The model is overfitting.

## How to Fix Underfitting?

- Use a more complex model  
- Add more features  
- Reduce regularization  

## How to Fix Overfitting?

- Use more training data  
- Apply regularization  
- Reduce model complexity  
- Use cross-validation  

## Bias vs Variance

Underfitting → High Bias  
Overfitting → High Variance  

Good models balance bias and variance.
